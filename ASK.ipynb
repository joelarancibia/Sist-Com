```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
señ_dig=[0,1,1,0,0,1,0,1,1,1,0,1] #señal digital
```

```python
num_mues=3  #numero de muestras
```

```python
señ_muest=[] #señal muestreada
```

```python
for i in range(1,12):
    u=np.ones(n_mue)
    m=u*señ_dig[i]
    señ_muest=np.concatenate((señ_muest,m))
```

```python
plt.subplot(3,1,1)
plt.plot(señ_muest)
```

```python
#señal portadora
```

```python
dim=len(señ_muest)
```

```python
t=np.linspace(0,5,dim)
```

```python
fp=5
```

```python
señ_p=np.cos(2*np.pi*fp*t)
```

```python
plt.subplot(3,1,2)
plt.plot(t,señ_p)
```

```python
#señal modulada
```

```python
señ_mod=señ_muest*señ_p
```

```python
plt.subplot(3,1,3)
plt.plot(t,señ_mod)
```

```python
plt.subplot(3,1,1)
plt.plot(señ_muest) #señal muestreada
plt.subplot(3,1,2)
plt.plot(t,señ_p) #señal portadora
plt.subplot(3,1,3)
plt.plot(t,señ_mod) #señal modulada
```
