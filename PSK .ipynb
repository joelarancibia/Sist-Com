```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
señal_digital = [0,1,1,0,0,1,0,1,1,1] #senial digital
```

```python
numero_muestras = 100 #numero de muestras
```

```python
señal_muestreada = []
```

```python
for i in range (1, 10):
    pulsos = np.ones(200) 
    muestras = pulsos * señal_digital[i]
    señal_muestreada = np.concatenate((señal_muestreada, muestras))
```

```python
plt.subplot(4,1,1)
plt.plot(señal_muestreada)
```

```python
dimension = len(señal_muestreada) #longitud del vector de la senial_muestreada
```

```python
t= np.linspace(0,10,dimension) #periodo de la portadora
```

```python
frec = 2 #frecuencia de la portadora
```

```python
portadora_seno = np.sin(2 * np.pi * frec * t)
```

```python
portadora_coseno = np.cos(2 * np.pi * frec *t)
```

```python
plt.subplot(4,1,2)
plt.plot(t, portadora_seno)
plt.subplot(4,1,3)
plt.plot(t, portadora_coseno)
```

```python
# senial modulada
```

```python
señal_modulada = []
```

```python
for i in range(0,dimension):
    if señal_muestreada[i] ==0:
        posicion_cero = np.array([portadora_seno[i]])
        señal_modulada = np.concatenate((señal_modulada, posicion_cero))
    else:
        posicion_uno = np.array([portadora_coseno[i]])
        señal_modulada = np.concatenate((señal_modulada, posicion_uno))
```

```python
plt.subplot(4,1,4)
plt.plot(t, señal_modulada)
```

```python
plt.subplot(4,1,1)
plt.plot(señal_muestreada)
plt.subplot(4,1,2)
plt.plot(t, portadora_seno)
plt.subplot(4,1,3)
plt.plot(t, portadora_coseno)
plt.subplot(4,1,4)
plt.plot(t, señal_modulada)
```
